## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
start = time()
kmeans.fit(df.iloc[:, 1:])
end = time()

kmeans_tt = end - start
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
df["Cluster"] = df["Private"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
df.head()
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
import sklearn.metrics as mm
```

```python
mm.confusion_matrix(df["Cluster"], kmeans.labels_)
```

```python
print(mm.classification_report(df["Cluster"], kmeans.labels_))
```

## Trying another algorithm

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:-1], df.iloc[:, -1], test_size = 0.3, random_state = 0)

start = time()
model = DecisionTreeRegressor().fit(X_train, y_train)
end = time()
DT_tt = end - start

pred = model.predict(X_test)
print(mm.classification_report(y_test, pred))
```

```python
def convert(array):
    values = []
    for i in array:
        if i == 1:
            values.append("Private")
        else:
            values.append("Public")
    return values
```

```python
pred_df = X_test.copy()
pred_df["Prediction"] = convert(pred)
pred_df["Real_Value"] = convert(y_test)
```

```python
pred_df
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 6))

axes[0].set_title("Training Time")
axes[0].set_ylabel("Time ms")
sns.barplot(x = [0, 1],
            y = [kmeans_tt / 1000, DT_tt / 1000],
            hue = ["KMeans", "DecsionTree"],
            ax = axes[0],
           palette = "Paired");

axes[1].set_title("Accuracy score")
axes[1].set_ylabel("Accuracy")
sns.barplot(x = [0, 1],
            y = [mm.accuracy_score(df["Cluster"], kmeans.labels_), mm.accuracy_score(y_test, pred)],
            hue = ["KMeans", "DecsionTree"],
            ax = axes[1],
           palette = "crest");


for i in axes:
    i.set_xticks([0, 1], ["KMeans", "DecsionTree"])
    i.set_xlabel("Algorithm")
```
